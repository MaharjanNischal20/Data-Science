```python
import pandas as pd
import numpy as np
```

```python
titanic = pd.read_csv('Titanic-Dataset.csv')
```

```python
titanic.head()
```

```python
titanic.info()
```

```python
titanic.describe()
```

```python
del titanic['Name']
titanic.head()
```

```python
del titanic['Ticket']
titanic.head()
```

```python
del titanic['Cabin']
titanic.head()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder() #Initializing an object of class LabelEncoder
titanic['Sex'] = labelencoder.fit_transform(titanic['Sex'])
```

```python
titanic.head()
```

```python
titanic.isnull().sum()
```

```python
#Fill the missing age of the survived one with the mean age of the survival people and non survive age with non survival age ko mean
meanS = titanic[titanic.Survived == 1].Age.mean()
meanN = titanic[titanic.Survived == 0].Age.mean()
```

```python
meanS
```

```python
meanN
```

```python
# age nabhako ma chai mean le replace hunxa ani age bhairako ma chai j ko tei aauxa
titanic['Age'] = np.where(pd.isnull(titanic.Age) & titanic['Survived'] == 1 , meanS , titanic['Age'])
titanic.head(17)
```

```python
# inplace le chai jun column herira ho tei ma operate garne
titanic.Age.fillna(meanN,inplace=True)
titanic.head()
```

```python
# yesma chai if any row ma null xa bhane tyo row hatauxa
titanic.dropna(inplace=True)
titanic.head()
```

```python
def getEmb(str):
    if str == "S":
        return 1
    elif str == 'O':
        return 2
    else:
        return 3
titanic['Embark'] = titanic['Embarked'].apply(getEmb)
titanic.head()
```

```python
titanic.isnull().sum()
```

```python
survivedQ = titanic[titanic.Embarked == 'Q'][titanic.Survived == 1].shape[0]
survivedS = titanic[titanic.Embarked == 'S'][titanic.Survived == 1].shape[0]
survivedC = titanic[titanic.Embarked == 'C'][titanic.Survived == 1].shape[0]
```

```python
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
males=(titanic['Sex']==1).sum()
females=(titanic['Sex']==0).sum()
print(males)
print(females)
p=[males,females]
plt.pie(p,#giving array
       labels=['males','females'],#corresponding giving lables
       colors=['green','yellow'],#coresponding color
       explode=(0.15,0),#how much the gap should me there between the pies
       startangle=0)#what start angle should be given 
plt.axis('equal')
plt.show()
```

```python
MaleS= titanic[titanic.Sex==1][titanic.Survived==1].shape[0]
print(MaleS)
MaleN= titanic[titanic.Sex==1][titanic.Survived==0].shape[0]
print(MaleN)
FemaleS= titanic[titanic.Sex==0][titanic.Survived==1].shape[0]
print(FemaleS)
FemaleN= titanic[titanic.Sex==0][titanic.Survived==0].shape[0]
print(MaleN)
```

```python
chart = [MaleS,MaleN,FemaleS,FemaleN]
colors=['lightskyblue','yellowgreen','Yellow','Orange']
labels=['Survived Male','Not Survived Male','Survived Female','Not Survived Female']
explode=[0,0.05,0,0.1]
plt.pie(chart,labels=labels,colors=colors,explode=explode,startangle=100,counterclock=False,autopct="%.2f%%")
plt.axis('equal')
plt.show()
```
